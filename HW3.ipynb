```python
def unblur(column, delim='-'):
    replacement_indexs=[]
    for i in range(len(column)):
        if type(column[i]) == float :
            x=np.random.random_sample()*100
            replacement_indexs.append(x)
        else:
            stringvalues=list(column[i])
            if len(stringvalues)==1:
                if stringvalues[0]=='.':
                    x=np.random.random_sample()*100
                    replacement_indexs.append(x)
                else:
                    replacement_indexs.append(int(stringvalues[0]))
            elif len(stringvalues)==2:
                x=np.random.normal(50, 10, 1)[0]
                replacement_indexs.append(x)
            elif len(stringvalues)==3:
                if stringvalues[0]=='G' and stringvalues[1]=='E':
                    value=int(stringvalues[2])
                    x=np.random.randint(value,100, size=1)[0]
                    replacement_indexs.append(x)
                if stringvalues[0]=='G' and stringvalues[1]=='T':
                    value=int(stringvalues[2])
                    x=np.random.randint(value+1,100, size=1)[0]
                    replacement_indexs.append(x)
                elif stringvalues[0]=='L' and stringvalues[1]=='E':
                    value=int(stringvalues[2])
                    x=np.random.randint(value, size=1)[0]
                    replacement_indexs.append(x)
                elif stringvalues[0]=='L' and stringvalues[1]=='T':
                    value=int(stringvalues[2])
                    x=np.random.randint(1,value-1, size=1)[0]
                    replacement_indexs.append(x)
                else:
                    x=np.random.randint(int(stringvalues[0]),int(stringvalues[2]), size=1)[0]
                    replacement_indexs.append(x)
            elif len(stringvalues)==4:
                if stringvalues[0]=='G' and stringvalues[1]=='E':
                    value=int(stringvalues[2]+stringvalues[3])
                    x=np.random.randint(value,100, size=1)[0]
                    replacement_indexs.append(x)
                if stringvalues[0]=='L' and stringvalues[1]=='E':
                    value=int(stringvalues[2]+stringvalues[3])
                    x=np.random.randint(value, size=1)[0]
                    replacement_indexs.append(x)
                if stringvalues[0]=='G' and stringvalues[1]=='T':
                    value=int(stringvalues[2]+stringvalues[3])
                    x=np.random.randint(value+1,100, size=1)[0]
                    replacement_indexs.append(x)
                if stringvalues[0]=='L' and stringvalues[1]=='T':
                    value=int(stringvalues[2]+stringvalues[3])
                    x=np.random.randint(1,value-1, size=1)[0]
                    replacement_indexs.append(x)
            elif len(stringvalues)==5:
                value1=int(stringvalues[0]+stringvalues[1])
                value2=int(stringvalues[3]+stringvalues[4])
                x=np.random.randint(value1,value2, size=1)[0]
                replacement_indexs.append(x)
            else:
                print(stringvalues)
  
    return(replacement_indexs)
            
            
```

```python
def budget_reducer(data):
    school_ids=[]
    amount_reduced=[]
    for i in range(data.shape[0]):
        amount=data.iloc[i]['diff']
        if amount>0:
            school_ids.append(data.iloc[i]['LEAID'])
            if data.iloc[i]['diff']<=data.iloc[i]['TFEDREV']:
                amount_reduced.append(amount)
            else:
                amount_reduced.append(data.iloc[i]['TFEDREV'])
    fed_budget=data['TFEDREV'].sum()
    reduced_amount=.15*fed_budget
    current_reduced=sum(amount_reduced)
    while reduced_amount < current_reduced:
        current_reduced=sum(amount_reduced)
        if current_reduced>reduced_amount:
            reducedeach=(current_reduced-reduced_amount)/len(amount_reduced)
        for x in range(len(amount_reduced)):
            if amount_reduced[x] >= reducedeach:
                amount_reduced[x]=amount_reduced[x]-reducedeach
            else:
                amount_reduced[x]=amount_reduced[x]
        current_reduced=sum(amount_reduced)
    print(sum(amount_reduced))
    school_red=pd.DataFrame([school_ids,amount_reduced]).transpose()
    school_red.columns=['LEAID',"AMOUNT reduced"]
    return(school_red)
```

        1.) Determine all districts that had a positive difference between their funding and their expenditure. 
        2.) Reduce district federal funding by the difference or by the federal funding whichever was smaller.
        3.) Determine whether the reduction was greater than needed. If so, iteratively reduce the reduction among all schools that had funding taken until the reduction equalled 15% of last year's budget.
        
This process was chose because it is clear, generally objective, and does not impact the learning of the schools that the money
